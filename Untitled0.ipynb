```python
x=np.arange(0,4*np.pi,0.1)
print(x)
y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)
print(y1,'\n',y2,'\n',y3)
fig=axx.figure(figsize=(10,4))
axx.subplot(1,3,1)
axx.title('sin(x)')
axx.plot(x,y1)

axx.subplot(1,3,2)
axx.title('cos(x)')
axx.plot(x,y2)

axx.subplot(1,3,3)
axx.title('tan(x)')
axx.plot(x,y3)
```

"""numpy.linspace(start,stop,num,endpoint,retstep,dtype)"""

start-The starting value of the sequence
stop-The end value of the sequence, included in the sequence if endpoint set to true
num-The 
retstep- If true, returns samples and step between the consequitive numbers
dtype- Data type of output ndarray

```python
comp=['Rel','TATA','Adani','Azhar']
rev1=[40,30,60,20]
rev=[110,90,100,200]
axx.xticks(xpos,comp)
axx.ylabel("revenue")
xpos=np.arange(len(comp))
print(xpos)
axx.bar(xpos,rev,label="revenue",width=0.2, color='black')    #prints the position of the first graph with the given width, label and color
axx.bar(xpos+0.2,rev1,label="money",width=0.2, color='red')   #prints the position of the second graph with the given width, label and color
axx.legend()
```

```python
exp_vals=[15000,600,400,400,250]
exp_labels=['home rent','food','internet','car','others']
axx.pie(exp_vals,labels=exp_labels,shadow=True)
axx.axis("equal")
axx.show()
```

```python
axx.axis("equal")
axx.pie(exp_vals,labels=exp_labels,shadow=True,autopct='%1.1f%%',radius=1.5)
axx.show()




# explode
exp_vals=[1500,600,400,400,250]
axx.axis("equal")
axx.pie(exp_vals,labels=exp_labels,shadow=True,autopct='%1.1f%%',radius=1.5,explode=[0,0.3,0,0.1,0.2])
axx.show()

```
